```python
years = [str(x) for x in np.arange(1950, 2010)]
```

```python
sf_years = {}
for year in years:
    sf_years[year] = sf_joined[sf_joined['GYear'] == year]
```

```python
cite_by_year = {}
for year in years:
    sf_1y = sf_years[year]
    cite_by_year[year] = sf_1y.groupby('Citation', gl.aggregate.COUNT).sort('Count', ascending=False)
```
