```python
def run_on_start_1():
    print "loading patent_matcher..."
    with open('../data/patent_matcher.pkl', 'rb') as handle:
        matcher = pkl.load(handle)
    return matcher

def run_on_start_2():
    print "loading dataframe..."
    with open('../data/patent_dataframe.pkl', 'rb') as handle:
        df = pkl.load(handle)
    return df
```

```python
citation_data = df_filtered['forward-citations_count']
citation_plot_url = []
for d in citation_data:
    url = plot_count(d)
    citation_plot_url.append(url)
```

```python
def plot_count(p, d):
    x = []
    y = []
    for k, v in d.iteritems():
        if v > 0:
            x.append(int(k))
            y.append(int(v))
#         x.append(int(k))
#         y.append(int(v) + 0.1)
    trace_one = Bar(
            x=x,
            y=y,
            name = 'US' + p,
            marker=Marker(
                color=a.next()
                )
            )
    
    return trace_one

```

```python
trace_list = []
citation_data = df_filtered['forward-citations_count']
a = color_generator()
for pat, cit in zip(pat_id, citation_data):
#     print pat, cit
    trace_one = plot_count(pat, cit)
    trace_list.append(trace_one)
```

```python
trace_list
```

```python
data = Data(trace_list)
layout = Layout(
    barmode='stack'
)
fig = Figure(data=data, layout=layout)
plot_url = py.iplot(fig, filename='stacked-bar')
```
