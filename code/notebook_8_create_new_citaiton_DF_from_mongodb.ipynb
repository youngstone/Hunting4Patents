```python
def func_clean_reference(citations):
    '''
    INPUT: an entry
    OUTPUT: an entry
    '''
    if citations:
        result = []
        for cite in citations:
            result.append("".join(re.findall('\d+', cite)))
        return result

def func_clean_class(classes):
    '''
    INPUT: an entry
    OUTPUT: an entry
    '''
    if classes:
        result = [cls.strip() for cls in classes.split(',')]
        return result

def func_clean_claim(claims):
    '''
    INPUT: an entry
    OUTPUT: an entry
    '''
    if claims:
        result = [clm.strip() for clm in claims.split('\n')][1:]
        return result
```

```python
df_raw_1['US Patent References'] = df_raw_1['US Patent References'].apply(func_clean_reference)
df_raw_1['Filing Date'] = pd.to_datetime(df_raw_1['Filing Date'])
df_raw_1['Other Classes'] = df_raw_1['Other Classes'].apply(func_clean_class)
df_raw_1['Primary Class'] = df_raw_1['Primary Class'].apply(func_clean_class)
df_raw_1['Claims'] = df_raw_1['Claims'].apply(func_clean_claim)
```

```python
def func_clean_backward_citation(citations):
    '''
    INPUT: an entry
    OUTPUT: an entry
    '''
    result = []
    if type(citations) != float:
        for i in xrange(len(citations)):
            result.append(citations[i]['Cited Patent'])
    return result
```

```python
def func_clean_forward_citation(citations):
    '''
    INPUT: an entry
    OUTPUT: an entry
    '''
    result = []
    if type(citations) != float:
        for i in xrange(len(citations)):
            result.append(citations[i]['Citing Patent'])
    return result
```

```python
df_raw_3['backward-citations'] = df_raw_3['backward-citations'].apply(func_clean_backward_citation)
```

```python
df_raw_3['forward-citations'] = df_raw_3['forward-citations'].apply(func_clean_forward_citation)
```

```python
df_raw_3
```

```python
def func_cite_by_examiner(citations):
    '''
    INPUT: an entry
    OUTPUT: an entry
    '''
    num_examiner = 0
    num_inventor = 0
    if len(citations) > 0:
        for cite in citations:
            if len(cite) != 0:
                if cite[-1] == '*':
                    num_examiner += 1
                else:
                    num_inventor += 1
    return num_examiner
```

```python
def func_cite_by_inventor(citations):
    '''
    INPUT: an entry
    OUTPUT: an entry
    '''
    num_examiner = 0
    num_inventor = 0
    if len(citations) > 0:
        for cite in citations:
            if len(cite) != 0:
                if cite[-1] == '*':
                    num_examiner += 1
                else:
                    num_inventor += 1
    return num_inventor
```

```python
df_raw_3['backward-citations-by-examiner'] = df_raw_3['backward-citations'].apply(func_cite_by_examiner)
df_raw_3['backward-citations-by-inventor'] = df_raw_3['backward-citations'].apply(func_cite_by_inventor)
df_raw_3['forward-citations-by-examiner'] = df_raw_3['forward-citations'].apply(func_cite_by_examiner)
df_raw_3['forward-citations-by-inventor'] = df_raw_3['forward-citations'].apply(func_cite_by_inventor)
```
