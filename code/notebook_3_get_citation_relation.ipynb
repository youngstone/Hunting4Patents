```python
import pandas as pd
import numpy as np
from datetime import date
import graphlab as gl
import networkx as nx
from collections import Counter, defaultdict
import matplotlib.pyplot as plt
import plotly.plotly as ply
%matplotlib inline
```

```python
df = pd.read_pickle('../data/dataframe.pkl')
```

```python
today = date.today() 
```

```python
screening_day = date(today.year - 20, today.month, today.day)
```

```python
df_select = df[df['Filing Date'] > screening_day]
```

```python
df_select.sort('Filing Date', axis=0).limit(5)
```

```python
df.sort('Filing Date', axis=0)
```

```python
df_citation = pd.DataFrame(columns=['patent', 'cited'])
```

```python
for ix in xrange(len(df)):
    pat = df['Number'].iloc[ix]
    cites = df['US Patent References'].iloc[ix]
    if cites:
        for cit in cites:
            df_citation = df_citation.append({'patent': pat, 'cited': cit}, ignore_index=True)
```

```python
df_citation.shape
```

```python
sf = gl.SFrame(data=df_citation)
```

```python
sf.groupby('cited', gl.aggregate.COUNT).sort('Count', ascending=False)
```

```python
gr = gl.SGraph(sf, vid_field='patent', src_field='patent', dst_field='cited')
gr = gr.add_edges(sf, src_field='patent', dst_field='cited')
```

```python
pr = gl.pagerank.create(gr)
```

```python
pr_out = pr['pagerank']
pr_out.sort('pagerank', ascending=False)[:10]
```
